```python
import pandas as pd
import numpy as np
from pathlib import Path
import re

# ---------- Paths ----------
INTERIM = Path(r"C:\Repositories\white-bowblis-nhmc\data\interim")
LITE_FP = INTERIM / "ccn_chow_lite.csv"
MCR_FP  = INTERIM / "mcr_chow_provider_events_all.csv"
OUT_FP  = INTERIM / "chow_facility_comparison.csv"

# ---------- Helpers ----------
def normalize_ccn_any(s: pd.Series) -> pd.Series:
    s = s.astype("string").fillna("").str.strip().str.upper()
    s = s.str.replace(r"[ \-\/\.]", "", regex=True)
    is_digits = s.str.fullmatch(r"\d+")
    return s.mask(is_digits, s.str.zfill(6)).replace({"": pd.NA})

def find_date_cols(df, patterns):
    cols = []
    for pat in patterns:
        cols += [c for c in df.columns if re.search(pat, c, re.I)]
    return sorted(set(cols))

def collapse_months(df, cols):
    """
    Return YYYY-MM pipe-joined list per row; empty string if no cols.
    Deduplicates by month, sorts ascending.
    """
    if not cols:
        return pd.Series([""] * len(df), index=df.index)
    dt = df[cols].apply(pd.to_datetime, errors="coerce")
    def row_to_str(r):
        months = sorted({pd.Period(d, "M") for d in r if pd.notna(d)})
        return "|".join(str(m) for m in months) if months else ""
    return dt.apply(row_to_str, axis=1)

def ensure_str_list(s):
    """Convert a pipe-delimited string (or NaN) to a list of YYYY-MM strings."""
    if not isinstance(s, str):
        return []
    if s.strip() == "":
        return []
    return [tok for tok in s.split("|") if tok]

def count_exact_overlap(a_str, b_str):
    A = set(ensure_str_list(a_str))
    B = set(ensure_str_list(b_str))
    return len(A & B)

def count_within_one_month(a_str, b_str):
    """Greedy match count where any A month within ±1 month of any B month."""
    A = [pd.Period(x, "M") for x in ensure_str_list(a_str)]
    B = [pd.Period(x, "M") for x in ensure_str_list(b_str)]
    used_b = set()
    matches = 0
    for a in A:
        for j, b in enumerate(B):
            if j in used_b:
                continue
            if abs((a - b).n) <= 1:
                used_b.add(j)
                matches += 1
                break
    return matches

def first_or_empty(s):
    L = ensure_str_list(s)
    return L[0] if L else ""

def last_or_empty(s):
    L = ensure_str_list(s)
    return L[-1] if L else ""

def classify_match(nl, nm, ld, md):
    nl = int(nl or 0); nm = int(nm or 0)
    if nl == 0 and nm == 0:
        return "both_zero"
    if nl > 0 and nm > 0:
        return "both_yes_same_months" if (ld or "") == (md or "") else "both_yes_different_months"
    if nl > 0 and nm == 0:
        return "only_lite"
    if nl == 0 and nm > 0:
        return "only_mcr"
    return "unknown"

# ---------- Load ----------
lite = pd.read_csv(LITE_FP, dtype={"cms_certification_number":"string"}, low_memory=False)
mcr  = pd.read_csv(MCR_FP,  dtype={"cms_certification_number":"string"}, low_memory=False)

# Normalize CCNs
lite["cms_certification_number"] = normalize_ccn_any(lite["cms_certification_number"])
mcr["cms_certification_number"]  = normalize_ccn_any(mcr["cms_certification_number"])

# Identify date columns
lite_date_cols = find_date_cols(lite, [r"^chow_date_\d+$", r"^chow_\d+_date$"])
mcr_date_cols  = find_date_cols(mcr,  [r"^chow_\d+_date$"])

# Build comparison frames (robust to missing columns)
if "num_chows" in lite.columns:
    nl = pd.to_numeric(lite["num_chows"], errors="coerce")
else:
    nl = lite[lite_date_cols].notna().sum(axis=1) if lite_date_cols else pd.Series(0, index=lite.index)
lite_comp = pd.DataFrame({
    "cms_certification_number": lite["cms_certification_number"],
    "num_chows_lite": nl.fillna(0).astype(int),
    "lite_dates": collapse_months(lite, lite_date_cols)
}).drop_duplicates("cms_certification_number")

if "n_chow" in mcr.columns:
    nm = pd.to_numeric(mcr["n_chow"], errors="coerce")
else:
    nm = mcr[mcr_date_cols].notna().sum(axis=1) if mcr_date_cols else pd.Series(0, index=mcr.index)
mcr_comp = pd.DataFrame({
    "cms_certification_number": mcr["cms_certification_number"],
    "num_chows_mcr": nm.fillna(0).astype(int),
    "mcr_dates": collapse_months(mcr, mcr_date_cols)
}).drop_duplicates("cms_certification_number")

# Merge
comp = lite_comp.merge(mcr_comp, on="cms_certification_number", how="outer")
comp["num_chows_lite"] = comp["num_chows_lite"].fillna(0).astype(int)
comp["num_chows_mcr"]  = comp["num_chows_mcr"].fillna(0).astype(int)

# Agreement & diagnostics
comp["match_type"]           = comp.apply(lambda r: classify_match(r["num_chows_lite"], r["num_chows_mcr"], r["lite_dates"], r["mcr_dates"]), axis=1)
comp["n_common_exact"]       = comp.apply(lambda r: count_exact_overlap(r["lite_dates"], r["mcr_dates"]), axis=1)
comp["n_common_within_1m"]   = comp.apply(lambda r: count_within_one_month(r["lite_dates"], r["mcr_dates"]), axis=1)
comp["first_lite_date"]      = comp["lite_dates"].apply(first_or_empty)
comp["last_lite_date"]       = comp["lite_dates"].apply(last_or_empty)
comp["first_mcr_date"]       = comp["mcr_dates"].apply(first_or_empty)
comp["last_mcr_date"]        = comp["mcr_dates"].apply(last_or_empty)
comp["discrepancy_flag"]     = comp["match_type"].isin(["only_lite","only_mcr","both_yes_different_months"]).astype(int)

# Order columns nicely
cols = [
    "cms_certification_number",
    "num_chows_lite","num_chows_mcr",
    "lite_dates","mcr_dates",
    "first_lite_date","first_mcr_date","last_lite_date","last_mcr_date",
    "n_common_exact","n_common_within_1m",
    "match_type","discrepancy_flag"
]
comp = comp[cols].sort_values("cms_certification_number").reset_index(drop=True)

# Save + quick summary
comp.to_csv(OUT_FP, index=False)
print(f"[saved] {OUT_FP}  rows={len(comp):,}  cols={len(comp.columns)}")
print("\n=== match_type counts ===")
print(comp["match_type"].value_counts().to_string())
print("\n=== head ===")
print(comp.head(10).to_string(index=False))
```

```python
import pandas as pd
import numpy as np
from pathlib import Path

INTERIM = Path(r"C:\Repositories\white-bowblis-nhmc\data\interim")
IN_FP   = INTERIM / "chow_facility_comparison.csv"
OUT_FP  = INTERIM / "chow_facility_comparison_plus.csv"

comp = pd.read_csv(IN_FP, dtype={"cms_certification_number":"string"})

def as_list(s):
    if not isinstance(s, str) or s.strip()=="":
        return []
    return [tok for tok in s.split("|") if tok]

def to_periods(s):
    return [pd.Period(x, "M") for x in as_list(s)]

def months_diff(a, b):
    """Signed months: a - b, using first dates only; NaN if missing either."""
    if not isinstance(a, str) or not isinstance(b, str) or a=="" or b=="":
        return np.nan
    pa, pb = pd.Period(a, "M"), pd.Period(b, "M")
    return (pa - pb).n

def jaccard(a_list, b_list):
    A, B = set(as_list(a_list)), set(as_list(b_list))
    if not A and not B:
        return 1.0
    if not A or not B:
        return 0.0
    return len(A & B) / len(A | B)

def within_k_months(a_list, b_list, k=0):
    """Return True if there is at least one pair within k months (k=0 → exact)."""
    A, B = to_periods(a_list), to_periods(b_list)
    if not A or not B:
        return False
    for a in A:
        for b in B:
            if abs((a - b).n) <= k:
                return True
    return False

def only_in_a(a_list, b_list):
    A, B = set(as_list(a_list)), set(as_list(b_list))
    return "|".join(sorted(A - B))

def only_in_b(a_list, b_list):
    A, B = set(as_list(a_list)), set(as_list(b_list))
    return "|".join(sorted(B - A))

# Add diagnostics
comp["delta_first_months"] = comp.apply(lambda r: months_diff(r["first_lite_date"], r["first_mcr_date"]), axis=1)
comp["within_0m"]          = comp.apply(lambda r: within_k_months(r["lite_dates"], r["mcr_dates"], k=0), axis=1)
comp["within_1m"]          = comp.apply(lambda r: within_k_months(r["lite_dates"], r["mcr_dates"], k=1), axis=1)
comp["jaccard_months"]     = comp.apply(lambda r: jaccard(r["lite_dates"], r["mcr_dates"]), axis=1)
comp["only_in_lite_dates"] = comp.apply(lambda r: only_in_a(r["lite_dates"], r["mcr_dates"]), axis=1)
comp["only_in_mcr_dates"]  = comp.apply(lambda r: only_in_b(r["lite_dates"], r["mcr_dates"]), axis=1)

# Save the enhanced table
comp.to_csv(OUT_FP, index=False)
print(f"[saved] {OUT_FP}  rows={len(comp):,}  cols={len(comp.columns)}")

# -------- Quick summary prints --------
print("\n=== match_type counts ===")
print(comp["match_type"].value_counts().to_string())

print("\n=== delta_first_months (first LITE minus first MCR) ===")
print(comp["delta_first_months"].describe(percentiles=[.01,.05,.25,.5,.75,.95,.99]))

print("\n=== Agreement flags ===")
print("within_0m (any exact month overlap):", int(comp["within_0m"].sum()))
print("within_1m (any overlap within ±1m): ", int(comp["within_1m"].sum()))

print("\n=== Top discrepancy examples (only_lite) ===")
print(comp.loc[comp["match_type"]=="only_lite",
              ["cms_certification_number","lite_dates","mcr_dates","only_in_lite_dates","only_in_mcr_dates"]]
        .head(10).to_string(index=False))

print("\n=== Top discrepancy examples (only_mcr) ===")
print(comp.loc[comp["match_type"]=="only_mcr",
              ["cms_certification_number","lite_dates","mcr_dates","only_in_lite_dates","only_in_mcr_dates"]]
        .head(10).to_string(index=False))

print("\n=== Both yes, different months (first 10) ===")
print(comp.loc[comp["match_type"]=="both_yes_different_months",
              ["cms_certification_number","lite_dates","mcr_dates","n_common_exact","n_common_within_1m",
               "delta_first_months","jaccard_months","only_in_lite_dates","only_in_mcr_dates"]]
        .head(10).to_string(index=False))
```

```python
import pandas as pd
from pathlib import Path

# Load
comp_fp = Path(r"C:\Repositories\white-bowblis-nhmc\data\interim\chow_facility_comparison_plus.csv")
comp = pd.read_csv(comp_fp, dtype={"cms_certification_number":"string"})

print(f"[loaded] {comp_fp} rows={len(comp):,} cols={len(comp.columns)}")

# --- 1. Match type breakdown ---
print("\n=== Match types (counts) ===")
print(comp["match_type"].value_counts(dropna=False).to_string())

# --- 2. Discrepancy stats ---
disc = comp[comp["match_type"] != "both_zero"].copy()

print("\n=== Discrepancy counts ===")
print(disc["match_type"].value_counts().to_string())

# How many months apart are first events, for facilities with both sources
delta = comp.loc[comp["delta_first_months"].notna(), "delta_first_months"]

print("\n=== Delta between first LITE and first MCR months ===")
print(delta.describe(percentiles=[0.01,0.05,0.25,0.5,0.75,0.95,0.99]))

# Histogram of deltas (rounded to years)
print("\n=== Histogram of delta_first_months (rounded to years) ===")
print((delta // 12).value_counts().sort_index().to_string())

# --- 3. Focus: only_mcr before Jan 2017 ---
only_mcr = comp[comp["match_type"]=="only_mcr"].copy()
only_mcr["mcr_dates_list"] = only_mcr["mcr_dates"].fillna("").str.split("|")

def has_pre2017(lst):
    for d in lst:
        if d and pd.to_datetime(d, errors="coerce") < pd.Timestamp("2017-01-01"):
            return True
    return False

only_mcr["any_pre2017"] = only_mcr["mcr_dates_list"].apply(has_pre2017)
n_pre2017 = only_mcr["any_pre2017"].sum()

print(f"\n=== Only MCR flagged (n={len(only_mcr)}) ===")
print(f"→ {n_pre2017} ({n_pre2017/len(only_mcr)*100:.1f}%) had at least one MCR CHOW before 2017-01")

# --- 4. Extra: event count distributions ---
print("\n=== Number of CHOWs distribution (LITE vs MCR) ===")
print("Lite:\n", comp["num_chows_lite"].value_counts().sort_index().head(20).to_string())
print("\nMCR:\n", comp["num_chows_mcr"].value_counts().sort_index().head(20).to_string())

# Optional: export a table of mismatches with their deltas for inspection
out_fp = comp_fp.with_name("chow_discrepancy_deltas.csv")
comp.loc[comp["match_type"].isin(["both_yes_different_months","only_lite","only_mcr"])].to_csv(out_fp, index=False)
print(f"\n[saved] discrepancy subset with deltas -> {out_fp}")
```
