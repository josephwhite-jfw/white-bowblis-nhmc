```python
import pandas as pd
import numpy as np
from pathlib import Path

# ---------- Paths ----------
INPUT_FP  = Path(r"C:\Repositories\white-bowblis-nhmc\data\clean\pbj_panel.csv")
OUTPUT_FP = Path(r"C:\Repositories\white-bowblis-nhmc\data\clean\analytical_panel.csv")

# ---------- Load ----------
df = pd.read_csv(INPUT_FP, low_memory=False)

# ---------- Normalize blanks to NaN ----------
df = df.replace(r"^\s*$", np.nan, regex=True)

# ---------- Drop rows with any NaN in HPPDs ----------
hppd_cols = ["rn_hppd", "lpn_hppd", "cna_hppd", "total_hppd"]
before = len(df)
df = df.dropna(subset=hppd_cols)
print(f"[filter] drop rows with NaN in any HPPD: {before:,} -> {len(df):,}")

# ---------- Drop rows where provider_resides_in_hospital == 1 ----------
if "provider_resides_in_hospital" in df.columns:
    before = len(df)
    df = df[df["provider_resides_in_hospital"] != 1]
    print(f"[filter] drop 'provider_resides_in_hospital'==1: {before:,} -> {len(df):,}")

# ---------- Save ----------
df.to_csv(OUTPUT_FP, index=False)
print(f"[done] saved cleaned panel → {OUTPUT_FP}")
```
