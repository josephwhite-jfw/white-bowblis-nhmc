```python
import pandas as pd
import numpy as np
from pathlib import Path

# ---------- Paths ----------
INPUT_FP  = Path(r"C:\Repositories\white-bowblis-nhmc\data\clean\pbj_panel.csv")
OUTPUT_FP = Path(r"C:\Repositories\white-bowblis-nhmc\data\clean\analytical_panel.csv")

# ---------- Load ----------
df = pd.read_csv(INPUT_FP, low_memory=False)

# ---------- Normalize blanks to NaN ----------
df = df.replace(r"^\s*$", np.nan, regex=True)

# ---------- Drop rows with any NaN in HPPDs ----------
hppd_cols = ["rn_hppd", "lpn_hppd", "cna_hppd", "total_hppd"]
before = len(df)
df = df.dropna(subset=hppd_cols)
print(f"[filter] drop rows with NaN in any HPPD: {before:,} -> {len(df):,}")

# ---------- Drop rows where provider_resides_in_hospital == 1 ----------
if "provider_resides_in_hospital" in df.columns:
    before = len(df)
    df = df[df["provider_resides_in_hospital"] != 1]
    print(f"[filter] drop 'provider_resides_in_hospital'==1: {before:,} -> {len(df):,}")

# ---------- Save ----------
df.to_csv(OUTPUT_FP, index=False)
print(f"[done] saved cleaned panel → {OUTPUT_FP}")
```

```python
import pandas as pd

# === Load your analytical panel ===
panel_fp = r"C:\Repositories\white-bowblis-nhmc\data\clean\analytical_panel.csv"
df = pd.read_csv(panel_fp)

# === Basic shape ===
print(f"[info] Rows: {len(df):,} | Columns: {df.shape[1]}")

# === Count missing values ===
na_counts = df.isna().sum().sort_values(ascending=False)
na_pct = (na_counts / len(df) * 100).round(2)

# === Combine into a single table ===
na_table = pd.DataFrame({
    "missing_count": na_counts,
    "missing_pct": na_pct
}).query("missing_count > 0")

# === Display summary ===
print("\n[Null summary — only variables with at least one missing value]")
display(na_table.head(30))  # show first 30 vars with missing values

# === Optional: export to CSV if you want a full record ===
# na_table.to_csv(r"C:\Repositories\white-bowblis-nhmc\outputs\null_summary.csv")
```
