```python
import os
from pathlib import Path
import pandas as pd

# === Configure paths ===
PROJECT_ROOT = Path.cwd()
while not (PROJECT_ROOT / "src").is_dir() and PROJECT_ROOT != PROJECT_ROOT.parent:
    PROJECT_ROOT = PROJECT_ROOT.parent

RAW_DIR  = Path(os.getenv("NH_DATA_DIR", PROJECT_ROOT / "data" / "raw")).resolve()
MCR_DIR  = RAW_DIR / "medicare-cost-reports"         # where your .sas7bdat / .xpt live
OUT_DIR  = PROJECT_ROOT / "data" / "interim" / "mcr_raw_csv"
OUT_DIR.mkdir(parents=True, exist_ok=True)

print(f"[paths] MCR_DIR={MCR_DIR}")
print(f"[out]   OUT_DIR={OUT_DIR}")

# === Detect reader ===
use_pyreadstat = False
try:
    import pyreadstat
    use_pyreadstat = True
    print("[read] Using pyreadstat for SAS/XPT")
except Exception:
    print("[read] pyreadstat not available → will use pandas.read_sas fallback")

# === Gather files ===
files = sorted(list(MCR_DIR.glob("*.sas7bdat")) + list(MCR_DIR.glob("*.xpt")))
if not files:
    raise FileNotFoundError(f"No SAS files found in {MCR_DIR}")

# === Convert each to CSV (verbatim columns/rows) ===
for fp in files:
    out_csv = OUT_DIR / (fp.stem + ".csv")
    print(f"[convert] {fp.name} → {out_csv.name}")

    if use_pyreadstat:
        if fp.suffix.lower() == ".sas7bdat":
            df, meta = pyreadstat.read_sas7bdat(fp, disable_datetime_conversion=0)
        else:  # .xpt
            df, meta = pyreadstat.read_xport(fp, disable_datetime_conversion=0)
    else:
        # pandas fallback (works for both sas7bdat/xport); may be slower/less precise on some dtypes
        df = pd.read_sas(fp, format="sas7bdat" if fp.suffix.lower()==".sas7bdat" else "xport", encoding="latin1")

    # Write CSV exactly as loaded (no transforms)
    # Use UTF-8 so Excel/others open cleanly; keep NA as empty cells.
    df.to_csv(out_csv, index=False, encoding="utf-8")

    # Optional: also save value-label metadata when using pyreadstat (commented out)
    # if use_pyreadstat:
    #     (OUT_DIR / f"{fp.stem}.columns.txt").write_text("\n".join(df.columns), encoding="utf-8")
    #     (OUT_DIR / f"{fp.stem}.meta.txt").write_text(str(meta), encoding="utf-8")

print("[done] CSV copies written.")
```
