```python
import pandas as pd
from pathlib import Path
from concurrent.futures import ThreadPoolExecutor, as_completed
import time
import re 

BASE = Path(r"C:\Users\wrthj\OneDrive\NursingHomeData\provider-info-files")
OUT_FP = BASE / "provider_resides_in_hospital_by_ccn.csv"

# Column aliases
CCN_ALIASES  = ["cms_certification_number", "ccn", "provider_number"]
HOSP_ALIASES = ["provider_resides_in_hospital", "resides_in_hospital", "inhosp"]

def pick_col(cols, options):
    for c in options:
        if c in cols:
            return c
    return None

def to_bool_series(s: pd.Series) -> pd.Series:
    # Fast vector map for common encodings
    m = s.astype(str).str.strip().str.lower()
    true_vals  = {"1","y","yes","true","t"}
    false_vals = {"0","n","no","false","f"}
    out = pd.Series(pd.NA, index=s.index, dtype="object")
    out = out.mask(m.isin(true_vals),  True)
    out = out.mask(m.isin(false_vals), False)
    # numeric 0/1
    num = pd.to_numeric(m, errors="coerce")
    out = out.mask(num == 1, True)
    out = out.mask(num == 0, False)
    return out

def read_if_has_cols(fp: Path):
    # 1) Peek header only (fast)
    hdr = pd.read_csv(fp, nrows=0)
    ccn_col  = pick_col(hdr.columns, CCN_ALIASES)
    hosp_col = pick_col(hdr.columns, HOSP_ALIASES)
    if ccn_col is None or hosp_col is None:
        return None  # skip quickly

    # 2) Read only needed columns (fast C engine)
    df = pd.read_csv(fp, usecols=[ccn_col, hosp_col], dtype={ccn_col: "string"})
    df.columns = ["cms_certification_number", "provider_resides_in_hospital"]

    # Normalize CCN digits and zero-pad
    df["cms_certification_number"] = (
        df["cms_certification_number"]
        .str.replace(r"\D", "", regex=True)
        .str.zfill(6)
    )

    # Normalize indicator
    df["provider_resides_in_hospital"] = to_bool_series(df["provider_resides_in_hospital"])

    return df

# Gather files
files = [p for p in BASE.glob("provider_info_*.csv") if re.match(r"provider_info_\d{2}_\d{4}\.csv$", p.name)]
files = sorted(files)
print(f"[scan] {len(files)} files")

t0 = time.time()

# Optional: set to >1 to parallelize (I/O-bound, threads help)
MAX_WORKERS = 8

dfs = []
skipped = 0
with ThreadPoolExecutor(max_workers=MAX_WORKERS) as ex:
    futs = {ex.submit(read_if_has_cols, fp): fp for fp in files}
    for fut in as_completed(futs):
        fp = futs[fut]
        try:
            df = fut.result()
            if df is None:
                skipped += 1
            else:
                dfs.append(df)
        except Exception as e:
            skipped += 1

elapsed = time.time() - t0
print(f"[read] gathered {len(dfs)} frames, skipped {skipped} files in {elapsed:.1f}s")

# Stack/sort
stacked = pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame(columns=["cms_certification_number","provider_resides_in_hospital"])
stacked = stacked.sort_values("cms_certification_number", kind="stable").reset_index(drop=True)
print(f"[stack] rows={len(stacked):,}  unique CCNs(raw)={stacked['cms_certification_number'].nunique()}")

# Deduplicate per CCN with priority True > False > NA (simple score)
score_map = {True: 2, False: 1}
tmp = stacked.assign(_score=stacked["provider_resides_in_hospital"].map(score_map).fillna(0).astype(int))
best = (
    tmp.sort_values(["cms_certification_number","_score"], ascending=[True, False], kind="stable")
       .drop_duplicates(subset=["cms_certification_number"], keep="first")
       .drop(columns=["_score"])
       .reset_index(drop=True)
       .sort_values("cms_certification_number", kind="stable")
)

# Save
best.to_csv(OUT_FP, index=False)
print(f"[save] {OUT_FP}  rows={len(best):,}")
print(best["provider_resides_in_hospital"].value_counts(dropna=False))
```
